```python
llm = OpenAI(temperature=0.6)

name_prompt = PromptTemplate(
    input_variables=['power'],
    template='I want to create a new superhero with the power of {power}. Suggest a badass name for him'
)

name_chain = LLMChain(llm=llm, prompt=name_prompt)

moves_prompt = PromptTemplate(
    input_variables=['name'],
    template="Suggest few super moves for {name}. Return it as a list."
)

moves_chain = LLMChain(llm=llm, prompt=moves_prompt)
```

```python
from langchain.chains import SimpleSequentialChain

chain = SimpleSequentialChain(chains=[name_chain, moves_chain])
response = chain.run('Invisibility')
print(response)
```

```python
llm = OpenAI(temperature=0.6)

name_prompt = PromptTemplate(
    input_variables=['power'],
    template='I want to create a new superhero with the power of {power}. Suggest a badass name for him'
)

name_chain = LLMChain(llm=llm, prompt=name_prompt, output_key='name')

moves_prompt = PromptTemplate(
    input_variables=['name'],
    template="Suggest few super moves for {name}. Return it as a list."
)

moves_chain = LLMChain(llm=llm, prompt=moves_prompt, output_key='super moves')
```

```python
from langchain.chains import SequentialChain

seq_chain = SequentialChain(
    chains=[name_chain, moves_chain],
    input_variables=['power'],
    output_variables=['name', 'super moves']
)

seq_chain({'power': 'Lightning'})
```

```python
from langchain.agents import AgentType, initialize_agent, load_tools
from langchain.llms import OpenAI

tools = load_tools(['wikipedia', 'llm-math'], llm=llm)

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

agent.run("When was Bruce Wayne born? What is his present age in 2024")
```
